```python
# importing dependent libraries and modules
import sklearn
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

```python
cancer = load_breast_cancer()

# convert the data into a numpy array
cancer_data = np.array(cancer.data)

# converting to pandas dataframe
df_cancer = pd.DataFrame(cancer_data, columns=cancer.feature_names)

# split the data
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2)
```

```python
df_cancer.head()
```

```python
# building and training the model
model = KNeighborsClassifier(n_neighbors = 5)
model.fit(X_train,y_train)
```

```python
# evaluate the model's accuracy]
acc_test=model.score(X_test,y_test)

acc_test
```

```python
 # Predict the data
predicted = model.predict(X_test)

condition = ['Benign', 'Malignant']

for i in range(len(predicted)):
    print('Predicted:', condition[predicted[i]], 'Actual:', condition[y_test[i]])
```
